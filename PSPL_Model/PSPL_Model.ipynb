```python
# Import the libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Calculate the magnification for a point source-point lens system.
def pspl_magnification(t, t0, tE, u0):

    # Calculate the projected separation of the source from the lens as a function of time
    u = np.sqrt(u0**2 + ((t - t0) / tE)**2)


    # Calculate the magnification factor based on the separation
    A = (u**2 + 2) / (u * np.sqrt(u**2 + 4))

    return A
```

```python
# The path to ulwdc1_025_W149.txt in Google Drive.
file_path_25 = '/content/drive/My Drive/Roman Simulations/ulwdc1_025_W149.txt'


# Define the names for the columns in the dataset.
column_names = ['BJD', 'Magnitude', 'Error']


# Read the data into a DataFrame
data_25 = pd.read_csv(file_path_25, delim_whitespace=True, header=None, names=column_names)

# Remove any rows with missing values
data_25 = data_25.dropna()

# Reset the index of the DataFrame
data_25 = data_25.reset_index(drop=True)


data_25
```

```python
# Plot Magnitude vs. Time (BJD)

plt.figure(figsize=(20, 6))
plt.errorbar(data_25['BJD'], data_25['Magnitude'], yerr=data_25['Error'], fmt='.', markersize=2, elinewidth=0.5)
plt.xlabel('BJD (Time)')
plt.ylabel('Magnitude')
plt.title('ulwdc1_025_W149')
plt.gca().invert_yaxis()  # Magnitude decreases as brightness increases
plt.show()
```

```python
# Set the x-axis limits to zoom in on the peak of the single lens event.



plt.figure(figsize=(20, 6))
plt.errorbar(data_25['BJD'], data_25['Magnitude'], yerr=data_25['Error'], fmt='.', markersize=2, elinewidth=0.5)
plt.xlabel('BJD (Time)')
plt.ylabel('Magnitude')
plt.title('ulwdc1_025_W149')
plt.gca().invert_yaxis()  # Magnitude decreases as brightness increases

plt.xlim(2.45950e6, 2.46e6) # Set the x-axis limit
plt.show()
```

```python
# Convert Magnitude to Flux using the zero-point magnitude.

# Define the zero-point magnitude for the W149 filter.
ZP = 25.0



data_25['Flux'] = 10**((ZP - data_25['Magnitude']) / 2.5)  # Magnitude to Flux conversion
data_25['Flux_Error'] = data_25['Error'] * data_25['Flux'] * np.log(10) / 2.5  # Calculate error in Flux
```

```python
# Plot Flux vs. Time (BJD)

plt.figure(figsize=(20, 6))
plt.errorbar(data_25['BJD'], data_25['Flux'], yerr=data_25['Flux_Error'], fmt='.', markersize=2, elinewidth=0.5)
plt.xlabel('BJD (Time)')
plt.ylabel('Flux')
plt.title('ulwdc1_025_W149')
plt.show()
```

```python
# Set the x-axis limits to zoom in on the peak of the single lens event.

plt.figure(figsize=(20, 6))
plt.errorbar(data_25['BJD'], data_25['Flux'], yerr=data_25['Flux_Error'], fmt='.', markersize=2, elinewidth=0.5)
plt.xlabel('BJD (Time)')
plt.ylabel('Flux')
plt.title('ulwdc1_025_W149')
plt.xlim(2.45950e6, 2.46e6) # Set the x-axis limit
plt.show()
```

```python
# Calculate magnification to fit the PSPL model to the light curve.



# Calculate the median flux
median_flux = data_25['Flux'].median()

# Compute the magnification as the ratio of flux to median flux
magnification = data_25['Flux'] / median_flux


# Calculate the magnification error
# The formula considers both the flux error and the difference between flux and its error
# This uses the min-max error propagation approach to estimate the uncertainty
magnification_err = (((data_25['Flux'] + data_25['Flux_Error']) - (data_25['Flux'] - data_25['Flux_Error'])) / median_flux) * 0.5



# Add computed magnification and its error to the DataFrame
data_25['magnification'] = magnification
data_25['magnification_error'] = magnification_err
```

```python
# Plot Magnification vs. Time (BJD)

plt.figure(figsize=(20, 6))
plt.errorbar(data_25['BJD'], data_25['magnification'], yerr= data_25['magnification_error'], fmt='.', markersize=2, elinewidth=0.5)
plt.xlabel('BJD (Time)')
plt.ylabel('Magnification')
plt.title('ulwdc1_025_W149')
plt.show()
```

```python
# Set the x-axis limits to zoom in on the peak of the single lens event.

plt.figure(figsize=(20, 6))
plt.errorbar(data_25['BJD'], data_25['magnification'], yerr= data_25['magnification_error'], fmt='.', markersize=2, elinewidth=0.5)
plt.xlabel('BJD (Time)')
plt.ylabel('Magnification')
plt.title('ulwdc1_025_W149')
plt.xlim(2.45950e6, 2.46e6) # Set the x-axis limit
plt.show()
```

```python
# Keep only the data close to the event.
data_25 = data_25.loc[(data_25['BJD'] <= 2.46e6) & (data_25['BJD'] >= 2.45950e6)]


# Rescale the time to avoid large numbers in the fitting process.
data_25.loc[:, 'BJD'] = data_25['BJD'] - 2.45e6


# plot the final data
plt.figure(figsize=(20, 6))
plt.errorbar(data_25['BJD'], data_25['magnification'], yerr= data_25['magnification_error'], fmt='.', markersize=2, elinewidth=0.5)
plt.xlabel('BJD (Time)')
plt.ylabel('Magnification')
plt.title('ulwdc1_025_W149')
plt.show()
```

**Reading the Initial Guesses for Fitting Parameters from the Simulation Info File:**

**The columns are:**

idx Sky position | Ds | DL ML | u0 alpha t0 tE rE thetaE piE rhos | MP a inc phase q s period | | F087s W149s | F087l W149l | fs0 fs1 | | chi2 chi2

```python
import pandas as pd

# Define the file path
file_path_info = '/content/drive/My Drive/Roman Simulations/master_file.txt'

# Read the file into a DataFrame
df = pd.read_csv(file_path_info, delimiter='\t', header=None, names=['Index', 'Data'])


# The parameters of 25th Lightcurve
df['Index'][25]
```

**u0 = 0.0926006**

**t0 = 1432.56871104**

**tE = 18.6718**

**fs = 0.0319219**

```python
from scipy.optimize import minimize
import warnings



# Suppress specific runtime warnings
warnings.simplefilter(action='ignore', category=RuntimeWarning)




# Define the PSPL model function
def fun(t, t0, tE, u0, fs):
    u = np.sqrt(u0**2 + ((t - t0) / tE)**2)
    A = (u**2 + 2) / (u * np.sqrt(u**2 + 4))
    F = fs * A + (1 - fs)
    return F


# Define the chi-squared function
def chi2(theta, t, f, f_err):
    t0, tE, u0, fs = theta
    model = fun(t, t0, tE, u0, fs)
    inv_sigma2 = 1.0 / (f_err**2)
    return np.sum((f - model)**2 * inv_sigma2)





# Initial guesses for the parameters
t0_init = 9650
tE_init = 18.5
u0_init = 0.09
fs_init = 0.03






# Define bounds for each parameter
bounds = [(None, None),  # t0 has no bounds
          (0, None),     # tE should be positive
          (0, None),     # u0 should be non-negative
          (0, 1)]        # fs should be between 0 and 1



# Perform the minimization
result = minimize(chi2, [t0_init, tE_init, u0_init, fs_init],
                  args=(data_25['BJD'].values, data_25['magnification'].values, data_25['magnification_error'].values),
                  method='L-BFGS-B', bounds=bounds)







# Check if the optimization was successful
if not result.success:
    print("Optimization failed:", result.message)

# Extract the best-fit parameters
t0_fit, tE_fit, u0_fit, fs_fit = result.x

# Print the best-fit parameters and final log-likelihood
print(f"Best-fit parameters:")
print(f"t0 = {t0_fit:.5f}")
print(f"tE = {tE_fit:.2f}")
print(f"u0 = {u0_fit:.5f}")
print(f"fs = {fs_fit:.2f}")
print(f"Final chi2: {result.fun:.2f}")

# Generate the model data with the best-fit parameters
model_flux = fun(data_25['BJD'].values, t0_fit, tE_fit, u0_fit, fs_fit)



# Plot the observed magnification and the best-fit model
plt.figure(figsize=(20, 6))
plt.errorbar(data_25['BJD'], data_25['magnification'], yerr=data_25['magnification_error'], fmt='.', markersize=4, elinewidth=0.5, label='Observed Magnification', alpha=0.1)
plt.plot(data_25['BJD'], model_flux,'-r', label='Best-Fit Model', linewidth=2)
plt.xlabel('BJD (Time)')
plt.ylabel('Magnification')
plt.legend()
plt.title('ulwdc1_025_W149')
plt.grid(True)
plt.show()
```
